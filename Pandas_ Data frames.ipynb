# Insertion and Deletion By Index/records

```python
imdb_rating=pd.read_csv(r'Downloads/classess/the_imdb_rating.csv')
```

```python
imdb_rating
```

```python
#suppose i have to add new rows and columns of data inside the file
imdb_rating1=pd.DataFrame({'star_rating':[9.0,8.9,8.8,8.7,8.6,8.5],'title':['Lucifer','The Mentalist','Strange','The Doctor','Special Ops','Venom'],'genre':['Horror','Crime','Drama','Crime','Crime','Drama'],'duration':[190,500,120,650,400,700],'year':[2001,2002,2003,2004,2005,2006]})
```

```python
imdb_rating1
```

```python
#Now we have to append the data of imdb_rating1 in imdb_rating.
imdb_rating.append(imdb_rating1)
```

```python
#now 1 problem is still there i.e, index point again start from 0. so for that
imdb_rating.append(imdb_rating1,ignore_index=True)
#now it will continue from last index point and will add the data from that index.
```

```python
#now if we see that after adding these data , can we look again on our previous data 
imdb_rating
#yes of course, bcz when append fn used it will create an object inside the file . not add properly
```

```python
# now if we use update fn . it will delete the data from first to number of rows of the data which are going to be update.
imdb_rating.update(imdb_rating1)
```

```python
# if we are updating new data it will be update by dropping from rows and update properly. it is not creating object , so we can't back up our previous dropped(deleted) data.
imdb_rating
```

```python
#if you want to delete the data by using it index point . in that case
imdb_rating.drop(imdb_rating.index[[2]])
# index 2's data will be deleted.
```

```python
#but by index point data will not be deleted properly. we have still that data.
imdb_rating
# but for rows it will not work
```

```python
# to See Sorting results:   suppose i want to see only those data which is less than 200
imdb_rating[imdb_rating.duration>200]
#now the data will be shown only who is less than 200. you can change the 200 and sort it as you want for you ease.
```

```python
#now the data will be shown only who is less than 200. you can change the 200 and sort it as you want for you ease.
imdb_rating[imdb_rating.duration>300]
```

```python
imdb_rating[imdb_rating.genre=='Crime']
```

```python
# if you want to check two condition at single time. then 
imdb_rating[(imdb_rating.genre=='Crime') & (imdb_rating.duration>500)]
# here we are using bitwise & operator to perform two task in single line
```

```python
#if there is thousands of data so you can't go one by one to change index name. so for that
imdb_rating.index=["movie_"+str(i) for i in range(1,imdb_rating["duration"].size+1)]
```

```python
#suppose we are having any file where some value are not given. so in python it will show like Nan. but isn't appropriate
imdb_rating.fillna(2)
```

```python
#it will take the upper one subsequent data.
imdb_rating.fillna(method="ffill")
```
