```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#in seaborn we don't use the format of the file
tips = sns.load_dataset("tips")
tips
```

# in Seaborn it is mandatory to provide x value , y value and data(through which file you are getting dataset)

```python
#difference between seaborn and matplob here are we don't have to give labels one by one. either you can pass it in paramter
sns.relplot(x = 'total_bill' , y = 'tip' , data = tips)
```

```python
#here hue is used. hue is upgraded version of legend in seaborn. it will show specific data given in parameter.
# understand how data showing in the graph. 
#the graph is showing the data of total_bill and tip. from 0 the points is getting high that mean the person's total_bill is higher than other and tip is higher than other.
#now the colored point rise means, person's total_bill and tip is high and he is smoker. as you can see smoker's value yes/no.
#so, these are the comparison graph of the data given 
sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'smoker' )
```

```python
#here data will be shown with two comparison. here style kwargs will change the point style as per comparison of data.
# Now the data is showing like circle and cross mark.
#the circle mark is showing that the total_bill, tip, not a smoker and had choosen lunch.
#and cross mark is showing that total_bill,tip, smoker, and had choosen dinner
sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'smoker' ,style = 'time')
```

```python
# now here we are using size kwarg with value size of the dataset, 
#it is differentiating the data by sizes of the circle as you can see in the graph
sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'smoker' ,style = 'time' , size = 'size')
```

```python
#if you want to separate the data into two part for more clearance . you can use it row or col. it will separate and draw two graph.
# now the data is like for first one is showing howmany have lunch with all other aspects and 2nd graph for howmany have dinner with other aspects too.
sns.relplot(x = 'total_bill' , y = 'tip' , data = tips , hue = 'smoker' ,style = 'size' , row = 'time')
```

style = shape(cross, circle,etc)

```python
# Using matplolib
import matplotlib.pyplot as plt
plt.scatter( x = "R&D Spend" , y = "Profit" , data = df)
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Spend VS Profit")
plt.show()
```
